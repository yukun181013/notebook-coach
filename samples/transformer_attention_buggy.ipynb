# Tiny attention walkthrough

This standard-library example computes attention scores and weights.

```python
import math
import random

def dot(left, right):
    return sum(a * b for a, b in zip(left, right))

def softmax(values):
    shifted = [math.exp(value - max(values)) for value in values]
    total = sum(shifted)
    return [value / total for value in shifted]
```

```python
queries = [[random.random() for _ in range(3)] for _ in range(2)]
keys = [[random.random() for _ in range(3)] for _ in range(3)]
```

```python
scale = math.sqrt(3)
scores = [[dot(query, key) / scale for key in keys] for query in queries]
```

```python
# Intentional teaching bug: normalize columns across queries.
normalized_columns = [softmax(list(column)) for column in zip(*scores)]
attention_weights = [list(row) for row in zip(*normalized_columns)]
attention_weights
```

Each key distributes attention across queries, so every key column should sum to one.
